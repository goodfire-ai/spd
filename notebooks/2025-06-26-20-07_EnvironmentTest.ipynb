```python
# Test plotting
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test Plot: Sine and Cosine Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
